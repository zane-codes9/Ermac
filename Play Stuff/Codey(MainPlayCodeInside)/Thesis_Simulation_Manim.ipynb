```python
from IPython.display import Video
from manim import *

# Write script to a file
code = """
from manim import *

class HeatCollapse(Scene):
    def construct(self):
        axes = Axes(
            x_range=[30, 52, 5],
            y_range=[0, 300, 50],
            axis_config={"include_numbers": True},
            x_length=7,
            y_length=5,
        ).add_coordinates()

        x_label = axes.get_x_axis_label("Temp (°C)", edge=RIGHT, direction=DOWN, buff=0.2)
        y_label = axes.get_y_axis_label("L(T)", edge=UP, direction=LEFT, buff=0.2) # This is to represent a leading indicator (deaths would be a lagging indicator)

        self.play(Create(axes), Write(x_label), Write(y_label))

        blue_graph = axes.plot(lambda x: 5 * (x - 30), color=BLUE, x_range=[30, 49.0])
        blue_label = Text("Linear Model", color=BLUE, font_size=24).next_to(blue_graph, UP).shift(LEFT)

        self.play(Create(blue_graph), Write(blue_label))
        self.wait(0.5)

        t_crit = 49.6
        red_graph = axes.plot(
            lambda x: 100 / (t_crit - x),
            color=RED,
            x_range=[30, 49.0]
        )

        red_label = Text("Non-Linear model), color=RED, font_size=24).move_to(axes.c2p(40, 80))

        wall = DashedLine(
            start=axes.c2p(t_crit, 0),
            end=axes.c2p(t_crit, 300),
            color=RED
        )
        wall_text = Text("49.6°C", color=RED, font_size=20).next_to(wall, UP)

        self.play(FadeOut(blue_graph), FadeOut(blue_label))
        self.play(Create(wall), Write(wall_text))
        self.play(Write(red_label))
        self.play(Create(red_graph), run_time=2.5)

        dot_lin = Dot(axes.c2p(49, 95), color=BLUE)
        dot_phys = Dot(axes.c2p(49, 166), color=RED)

        arrow = Arrow(dot_lin.get_center(), dot_phys.get_center(), buff=0, color=YELLOW)
        gap_text = MathTex(r"\Delta", font_size=34, color=YELLOW).next_to(arrow, RIGHT)

        self.play(FadeIn(dot_lin), FadeIn(dot_phys))
        self.play(GrowArrow(arrow), Write(gap_text))

        self.wait(2)
"""

with open("heat_script.py", "w") as f:
    f.write(code)

!manim -qm -v WARNING heat_script.py HeatCollapse

print("sIMUlation:")
Video("media/videos/heat_script/720p30/HeatCollapse.mp4", embed=True, html_attributes="controls autoplay loop")
```

```python
from IPython.display import Video
from manim import *

code = """
from manim import *

class HeatCollapse(Scene):
    def construct(self):
        axes = Axes(
            x_range=[30, 52, 5],
            y_range=[0, 300, 50],
            axis_config={"include_numbers": True},
            x_length=7,
            y_length=5,
        ).add_coordinates()

        # We use edge (where on line) and direction (where relative to that point)
        x_label = axes.get_x_axis_label("Temp", edge=RIGHT, direction=DOWN, buff=0.7)
        y_label = axes.get_y_axis_label("Stress", edge=UP, direction=LEFT, buff=0.2)

        self.play(Create(axes), Write(x_label), Write(y_label), run_time=2) # Slowed down setup


        # 2. LINEAR MODEL
        blue_graph = axes.plot(lambda x: 5 * (x - 30), color=BLUE, x_range=[30, 49.0])

        # Placing text using c2p (Coordinates to Point) gives you total control
        # Placing at Temp=38, Load=60
        blue_label = Text("Linear Model", color=BLUE, font_size=24).move_to(axes.c2p(40, 150))

        self.play(Write(blue_label))               # 1. Write the text
        self.play(Create(blue_graph), run_time=3)  # 2. Then draw the graph slowly

        # 3. ASYMPTOTIC MODEL
        t_crit = 49.6
        red_graph = axes.plot(
            lambda x: 100 / (t_crit - x),
            color=RED,
            x_range=[30, 49.0]
        )

        red_label = Text("Non-Linear Model", color=RED, font_size=24).move_to(axes.c2p(40, 150))

        wall = DashedLine(
            start=axes.c2p(t_crit, 0),
            end=axes.c2p(t_crit, 300),
            color=RED
        )
        wall_text = Text("49.6°C", color=RED, font_size=20).next_to(wall, UP)

        self.play(FadeOut(blue_graph), FadeOut(blue_label))
        self.play(Create(wall), Write(wall_text))

        self.play(Write(red_label))
        self.play(Create(red_graph), run_time=3) # Slowed down drawing

        # 4. THE DELTA
        dot_lin = Dot(axes.c2p(49, 95), color=BLUE)
        dot_phys = Dot(axes.c2p(49, 166), color=RED)

        arrow = Arrow(dot_lin.get_center(), dot_phys.get_center(), buff=0, color=YELLOW)

        # --- LESSON 3: DELTA SYMBOL ---
        # MathTex allows LaTeX symbols like \Delta
        gap_text = MathTex(r"\Delta", font_size=34, color=YELLOW).next_to(arrow, RIGHT)

        self.play(FadeIn(dot_lin), FadeIn(dot_phys))
        self.play(GrowArrow(arrow), Write(gap_text))

        self.wait(2)
"""

with open("heat_script.py", "w") as f:
    f.write(code)

!manim -qm -v WARNING heat_script.py HeatCollapse
Video("media/videos/heat_script/720p30/HeatCollapse.mp4", embed=True, html_attributes="controls autoplay loop")
```
